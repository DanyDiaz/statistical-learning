### Funciones para "escalar" información

```python
# Normalizará la variable: x_norm = (x - media(x))/(max(x)-(min(x)))
def normalizar(x, mediaEntrenamiento = None, maximoEntrenamiento = None, minimoEntrenamiento = None):
    if(mediaEntrenamiento == None or maximoEntrenamiento == None or minimoEntrenamiento == None):
        # Si no se reciben datos de entrenamiento, se calculan de los features "x" recibidos
        media = np.mean(x)
        maximum = np.amax(x)
        minimum = np.amin(x)
    else:
        media = mediaEntrenamiento
        maximum = maximoEntrenamiento
        minimum = minimoEntrenamiento
    
    return (x - media) / (maximum - minimum)

# Estandarizará la variable: x_stand = (x - media(x))/(desviacion(x)))
def estandarizar(x, mediaEntrenamiento = None, desviacionEntrenamiento = None):
    if(mediaEntrenamiento == None or desviacionEntrenamiento == None):
        # Si no se reciben datos de entrenamiento, se calculan de los features "x" recibidos
        media = np.mean(x)
        desviacion = np.std(x)
    else:
        media = mediaEntrenamiento
        desviacion = desviacionEntrenamiento
    
    return (x - media) / (desviacion)
```

### Regresión Softmax

Crear funciones para modelo de regresión Softmax:

```python
def calcularLogits(parametros, matrizX):
    return tf.matmul(matrizX, parametros)

def calcularSoftmax(logits):
    return tf.nn.softmax(logits)

def calcularPorcentajeCorrecto(logits, etiquetasY):
    softmaxY = calcularSoftmax(logits)
    boolSonIguales = tf.equal(tf.argmax(softmaxY, 1), tf.argmax(etiquetasY, 1))
    intSonIguales = tf.dtypes.cast(boolSonIguales, tf.int32)
    porcentajeCorrecto = tf.divide(tf.reduce_sum(intSonIguales), tf.shape(etiquetasY)[0])
    return porcentajeCorrecto

def agregarColumnaUnos(x):
    if len(x.shape) < 2:
        x = np.array(x, dtype=np.float32).reshape(x.shape[0], -1)
    
    return np.append(x, np.ones(x.shape[0]).reshape(-1,1), 1)

# Función que convertirá etiquetas, aplicando el algoritmo de one hot encoding, a una matriz de ceros y unos
# Suposiciones, las etiquetas serán números enteros desde 0 hasta un número positivo
def generarOneHotEncoding(etiquetas):
    etiquetasUnicas = np.unique(etiquetas)
    oneHot = np.zeros(shape=[etiquetas.shape[0], etiquetasUnicas.shape[0]], dtype=np.float32)
    filas = np.arange(etiquetas.shape[0])
    columnas = []
    for etiqueta in etiquetas:
        columnas.append(np.where(etiquetasUnicas == etiqueta)[0][0])
        
    oneHot[filas, columnas] = 1
    return oneHot
```

### Definir entrenamiento con Gradient Descent

```python
def trainingPorMiniBatchGradientDescent(trainX, trainY, testX, testY, epochs, learningRate, batchSize = -1):
    # Agregar columna de sesgo (con unos)
    trainX = agregarColumnaUnos(trainX)
    # Crear matriz con el algoritmo de one hot encoding
    yOneHot = generarOneHotEncoding(trainY)
    # Cantidad de datos, características y etiquetas enviadas
    cantidadDatos = trainX.shape[0]
    numCaracteristicas = trainX.shape[1]
    numEtiquetas = yOneHot.shape[1]
    # Parámetros
    parametros = tf.Variable(np.zeros((numCaracteristicas, numEtiquetas), dtype=np.float32), name="Parametros")
    # Dar formato adecuado a información de validación
    testX = agregarColumnaUnos(testX)
    # Crear matriz con el algoritmo de one hot encoding
    yTestOneHot = generarOneHotEncoding(testY)
    
    # Si el batch size es mayor que la cantidad de datos, simplemente se usa la cantidad de datos
    if batchSize > cantidadDatos | batchSize < 0:
        batchSize = cantidadDatos
    # Se determina número de iteraciones
    totalIteraciones = cantidadDatos // batchSize
    
    directorioModelo = './summaries/' + dt.datetime.now().strftime("%Y%m%d-%H%M%S") +  \
         '_RegresionLogisticaSoftmaxMiniBatchGradientDescent' \
        + '_epochs=' + str(epochs) + '_lr=' + str(learningRate) + '_batchSize=' + str(batchSize)

    writer = tf.summary.create_file_writer(directorioModelo)
    with writer.as_default():
        for epoch in range(epochs):
            for batch in range(totalIteraciones):
                inicioBatch = batch * batchSize
                finBatch = inicioBatch + batchSize - 1
                if (cantidadDatos - finBatch) < batchSize:
                    finBatch = cantidadDatos
                miniBatchX =  np.array(trainX[inicioBatch:finBatch,:], dtype=np.float32)
                miniBatchY = np.array(yOneHot[inicioBatch:finBatch,:], dtype=np.float32)
                
                with tf.GradientTape() as grad_tape:
                    grad_tape.watch(parametros)
                    with tf.name_scope("Prediccion"):
                        logits = calcularLogits(parametros, miniBatchX)
                    with tf.name_scope("Error"):
                        error = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(miniBatchY, logits))
                
                with tf.name_scope("Gradiente"):
                    gradiente = grad_tape.gradient(error, parametros)
                
                with tf.name_scope("Actualizar_Parametros"):
                    parametros.assign(parametros - learningRate * gradiente)
                    
            # Calcular error general de modelo en data set de entrenamiento
            logits = calcularLogits(parametros, np.array(trainX, dtype=np.float32))
            trainError = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(np.array(yOneHot, dtype=np.float32), logits))
            porcentajeAciertoTrain = calcularPorcentajeCorrecto(logits, np.array(yOneHot, dtype=np.float32))
            # Calcular error general de modelo en data set de test
            logits = calcularLogits(parametros, np.array(testX, dtype=np.float32))
            testError = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(np.array(yTestOneHot, dtype=np.float32), logits))
            porcentajeAciertoTest = calcularPorcentajeCorrecto(logits, np.array(yTestOneHot, dtype=np.float32))
            
            # Añadir errores a summary de tensorboard
            tf.summary.scalar("Train_Error", trainError, epoch)
            tf.summary.scalar("Train_Accuracy_Percentage", porcentajeAciertoTrain, epoch)
            tf.summary.scalar("Test_Error", testError, epoch)
            tf.summary.scalar("Test_Accuracy_Percentage", porcentajeAciertoTest, epoch)
            writer.flush()
            
    writer.close()

    print('Modelo MiniBatch. Epochs: ', epochs, ', learning rate: ', learningRate, ', Batch size: ', batchSize, \
          ', Error entrenamiento: ', trainError.numpy(), ', %Entrenamiento: ', porcentajeAciertoTrain.numpy(), \
         ', Error validación: ', testError.numpy(), ', %Validación: ', porcentajeAciertoTest.numpy())

    return parametros.numpy()
```

### Crear función para predicción de prendas

#### 1. Obtener parámetros de predicción

Se obtendrán los parámetros para predicción para los datos originales como para los datos normalizados y estandarizados:

```python
parametrosParaDatosOriginales = trainingPorMiniBatchGradientDescent(xFilteredTrainImages, filteredTrainLabels, xFilteredTestImages, \
                                                filteredTestLabels, 250, 0.000001, 256)
parametrosParaDatosNormalizados = trainingPorMiniBatchGradientDescent(xNormalizedTrainImages, filteredTrainLabels, xNormalizedTestImages, \
                                                filteredTestLabels, 250, 0.001, 32)
parametrosParaDatosEstandarizados = trainingPorMiniBatchGradientDescent(xStandarizedTrainImages, filteredTrainLabels, xStandarizedTestImages, \
                                                filteredTestLabels, 250, 0.001, 32)
```

#### 2. Creación de función

```python
def softmax(x):
    f_x = np.exp(x) / np.sum(np.exp(x))
    return f_x
```

```python
arr_etiquetas = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
def prediccionImagen(parametros, imagen, escalado=None, media = None, maximo = None, minimo = None, desviacion = None):
    imagenModificada = np.array(imagen, dtype=np.float32).reshape(-1, imagen.shape[0] * imagen.shape[1])
    imagenModificada = np.append(imagenModificada, np.ones(imagenModificada.shape[0]).reshape(1,-1), 1)
    if(escalado == 'normalizar'):
        comentario = '(normalizado)'
        imagenModificada = normalizar(imagenModificada, mediaEntrenamiento=media, maximoEntrenamiento=maximo, minimoEntrenamiento=minimo)
    elif(escalado == 'estandarizar'):
        comentario = '(estandarizado)'
        imagenModificada = estandarizar(imagenModificada, mediaEntrenamiento=media, desviacionEntrenamiento=desviacion)
    else:
        comentario = ''
    prediccion = np.matmul(imagenModificada, parametros)
    prediccion = softmax(prediccion)
    indicePrediccion = prediccion.argmax()
    if(indicePrediccion == 0):
        indiceEtiqueta = 1
    elif(indicePrediccion == 1):
        indiceEtiqueta = 3
    elif(indicePrediccion == 2):
        indiceEtiqueta = 5
    titulo = str(arr_etiquetas[indiceEtiqueta]) + ', probabilidad: ' + str(round(100 * prediccion[0,indicePrediccion], 2)) + '% '+ comentario
    plt.figure()
    plt.imshow(imagen, cmap='gray')
    plt.title(titulo)
    plt.colorbar()
```

#### 3. Mostrar 10 predicciones al azar de los datos de verificación

```python
import random
random.seed(2022)
randomIndices = random.sample(range(0, filteredTestImages.shape[0]), 10)

# Con datos originales
for indice in randomIndices:
    prediccionImagen(parametrosParaDatosOriginales, filteredTestImages[indice])
    
# Con datos normalizados
for indice in randomIndices:
    prediccionImagen(parametrosParaDatosNormalizados, filteredTestImages[indice], escalado='normalizar', \
                     media=trainMedia, maximo=trainMaximum, minimo=trainMinimum)
    
# Con datos estandarizados
for indice in randomIndices:
    prediccionImagen(parametrosParaDatosEstandarizados, filteredTestImages[indice], escalado='estandarizar', \
                     media=trainMedia, desviacion=trainSdt)
```
